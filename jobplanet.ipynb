```python
df3 = df2[["name", "title",  "recruitment_text","end_at", "skills", "primary_responsibility", "introduction", "required_qualification", "preferred_skill", "benefit"]]
```

```python
df3.columns = ['company_name', 'title', 'career', 'date', 'tech_stack', 'description', 'company_description', 'required', 'preferred_skill', 'welfare']
```

```python
df3
```

```python
df3.info()
```

```python
import numpy as np
df3['company_description'] = df3['company_description'].replace('', np.nan)
df3['company_description'] = df3['company_description'].replace(np.nan, None)


df3['preferred_skill'] = df3['preferred_skill'].replace('', np.nan)
df3['preferred_skill'] = df3['preferred_skill'].replace(np.nan, None)


df3['welfare'] = df3['welfare'].replace('', np.nan)
df3['welfare'] = df3['welfare'].replace(np.nan, None)
```

```python
pd.set_option('display.max_rows', 15)  # 다 보고싶으면 rows->columns
```

```python
import pandas as pd
```

```python
df3['date']
```

```python
# career list에서 첫번째 문자만 추출하기
df3['career'] = df3['career'].apply(lambda x: x[0].split(' ~ ')[0])
```

```python
df3.loc[df3['career'].str.contains('경력무관'), 'career'] = '-1'
df3.loc[df3['career'].str.contains('신입'), 'career'] = '0'     
```

```python
# 'tech_stack' 열의 각 리스트를 문자열로 변환하는 함수 정의
def list_to_string(lst):
    return ', '.join(lst)

# 'tech_stack' 열의 모든 요소를 문자열로 변환
df3['tech_stack'] = df3['tech_stack'].apply(list_to_string)
```

```python
#date 열의 형식 "yyyy.mm.dd"에서 "yyyy-mm-dd"로 변경

from datetime import datetime
df3['date'] = pd.to_datetime(df3['date'], format='%Y.%m.%d')
```

```python
# job_site_code 컬럼 추가, value = 1
```

```python
df3['job_site_code'] = '1'
```

```python
#required, preferred_skill, welfare 컬럼 합치기

df3['combined'] = df3['required'] + ', ' + df3['preferred_skill'] + ', ' + df3['welfare']
```

```python
import numpy as np
df3['combined'] = df3['combined'].replace(np.nan, None)
```

```python
df3.drop(columns=['required', 'preferred_skill', 'welfare'], axis = 1, inplace = True)
```

```python
jobplanet_final = df3.rename(columns = {'combined' : 'required'})
```

```python
jobplanet_final.columns
```

```python
jobplanet_final = jobplanet_final[['job_site_code', 'company_name', 'title', 'career', 'date', 'tech_stack', 'description',
       'company_description', 'required']]
```

```python
jobplanet_final
```

```python
jobplanet_final['required'] = jobplanet_final['required'].apply(lambda x: x.encode('cp949', 'ignore').decode('cp949') if x else None)
jobplanet_final['company_description'] = jobplanet_final['company_description'].apply(lambda x: x.encode('cp949', 'ignore').decode('cp949') if x else None)
jobplanet_final['description'] = jobplanet_final['description'].apply(lambda x: x.encode('cp949', 'ignore').decode('cp949') if x else None)
```
